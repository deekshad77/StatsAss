```python
crime_Data=pd.read_csv('crime_data.csv')
```

```python
crime_Data
```

```python
def norm(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
crime_Data_norm=norm(crime_Data.iloc[:,1:])
```

```python
crime_Data_norm
```

```python
dendrogram=sch.dendrogram(sch.linkage(crime_Data_norm,method='average'))
```

```python
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc=hc.fit_predict(crime_Data_norm)
```

```python
y_hc
```

```python
crime_Data['h_clusterid']=hc.labels_
```
